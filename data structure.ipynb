2. Explain the difference between mutable and immutable data types with
   examples.

-> # Mutable Data Types
- Can be modified after creation
- Changes made to the original data affect all references to it

Examples:

- Lists: my_list = [1, 2, 3]; my_list[0] = 10; print(my_list) # Output: [10, 2,
   3]
- Dictionaries: my_dict = {'a': 1}; my_dict['a'] = 2; print(my_dict) # Output:
  {'a': 2}
- Sets: my_set = {1, 2, 3}; my_set.add(4); print(my_set) # Output: {1, 2, 3, 4}

-> Immutable Data Types
- Cannot be modified after creation
- Changes made to the original data create a new copy

Examples:

- Integers: my_int = 5; my_int += 1; print(my_int) # Output: 6 (creates a new
   integer object)
- Floats: my_float = 3.14; my_float += 1.0; print(my_float) # Output: 4.14      (creates a new float object)
- Strings: my_str = 'hello'; my_str += ' world'; print(my_str) # Output: 'hello
  world' (creates a new string object)
- Tuples: my_tuple = (1, 2, 3); my_tuple += (4, 5); print(my_tuple) # Output:
  (1, 2, 3, 4, 5) (creates a new tuple object)

4. Describe how dictionaries store data.
-> Dictionaries store data as key-value pairs using a hashing mechanism. Each key is hashed, and the resulting hash determines the index in memory where the value is stored.


5. Why might you use a set instead of a list in Python?
